```python
import numpy as np

n = int(input("Nhập kích thước của vector a và b: "))
a = np.array([int(i) for i in input("Nhập mảng a: ").split()])
b = np.array([int(i) for i in input("Nhập mảng b: ").split()])

print("Vector a:")
print("Max:", np.max(a))
print("Min:", np.min(a))
print("Giá trị trung bình:", np.mean(a))
print("Độ lệch chuẩn:", np.std(a))
print("Giá trị trung vị:", np.median(a))

c = np.vstack((a, b)).T
print("\nMa trận c:")
print(c)

mean_b = np.mean(b)
std_b = np.std(b)
d = np.random.normal(mean_b, std_b, (n, n))
print("\nMa trận d (phân phối Gauss):")
print(d)

sum_cd = c + d
diff_cd = c - d
prod_cd = np.matmul(c, d)
inv_d = np.linalg.inv(d)
prod_transpose_inverse = np.matmul(d.T, inv_d)

print("\nTổng của ma trận c và d:")
print(sum_cd)
print("\nHiệu của ma trận c và d:")
print(diff_cd)
print("\nTích của ma trận c và d:")
print(prod_cd)
print("\nTích của ma trận chuyển vị của d và ma trận nghịch đảo của d:")
print(prod_transpose_inverse)

e = np.expand_dims(c, axis=0)
print("\nTensor numpy e:")
print(e)
```

```python
import numpy as np

diem = np.random.uniform(0, 10, (20, 3))

print("Điểm lớn nhất mỗi môn:", np.max(diem, axis=0))
print("\nĐiểm nhỏ nhất mỗi môn:", np.min(diem, axis=0))

diem = diem.ravel()
print("Mảng sau khi xóa vị trí điểm bằng 0:")
diem = np.delete(diem, np.where(diem == 0))
print(diem)

diem.sort(kind='quicksort')
print("\nMảng sau khi sắp xếp giảm dần:")
print(diem)

k = float(input("Nhập số thực k: "))
insert_index = np.searchsorted(diem, k, side='right')
print("\nVị trí chèn số k mà không phá vỡ tính sắp xếp:", insert_index)

diem = np.insert(diem, insert_index, k)
print("Mảng sau khi thêm số k:")
print(diem)
```
